```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from analysis_utils import box_plot_steps, box_plot_cpad

sns.set(style="white", rc={"grid.color": "lightgray"})
plt.rcParams["figure.dpi"] = 150
```

```python
ENV_COLORS = {
    "Baseline": "y",
    "Enhanced": "b",
    "Obstacle": "g",
    "FrameSkip": "g",
    "CAPS": "c",
}
```

## Part 1

```python
df_baseline = pd.read_csv("Baseline-v0.csv")
df_baseline["Environment"] = "Baseline"
df_enhanced_v0 = pd.read_csv("Enhanced-v0.csv")
df_enhanced_v0["Environment"] = "Enhanced"
df_enhanced_v1 = pd.read_csv("Enhanced-v1.csv")
df_enhanced_v1["Environment"] = "FrameSkip"
df_enhanced_caps = pd.read_csv("Enhanced-v1-caps.csv")
df_enhanced_caps["Environment"] = "CAPS"
```

```python
# Drop lines where "Episode Length" == 1
df_baseline = df_baseline[df_baseline["Episode Length"] != 1]
df_enhanced_v0 = df_enhanced_v0[df_enhanced_v0["Episode Length"] != 1]
df_enhanced_v1 = df_enhanced_v1[df_enhanced_v1["Episode Length"] != 1]
df_enhanced_caps = df_enhanced_caps[df_enhanced_caps["Episode Length"] != 1]
```

```python
box_plot_steps([df_baseline, df_enhanced_v0, df_enhanced_v1, df_enhanced_caps], ENV_COLORS, "0")
box_plot_cpad([df_baseline, df_enhanced_v0, df_enhanced_v1, df_enhanced_caps], ENV_COLORS, "0")
```

## Part 2

```python
df_obstacles = pd.read_csv("Obstacle-v0.csv")
df_obstacles["Environment"] = "Obstacle"
df_obstacles_caps = pd.read_csv("Obstacle-v0-caps.csv")
df_obstacles_caps["Environment"] = "CAPS"
df_obstacles = df_obstacles[df_obstacles["Episode Length"] != 1]
df_obstacles_caps = df_obstacles_caps[df_obstacles_caps["Episode Length"] != 1]
box_plot_steps([df_obstacles, df_obstacles_caps], ENV_COLORS, "1")
box_plot_cpad([df_obstacles, df_obstacles_caps], ENV_COLORS, "1")
```
