```python
import pandas as pd
import numpy as np
```

```python
df_baseline = pd.read_csv("Baseline-v0.csv")
df_baseline["Environment"] = "Baseline"
df_baseline_caps = pd.read_csv("Baseline-v0-caps.csv")
df_baseline_caps["Environment"] = "Baseline-CAPS"

df_baseline_fs = pd.read_csv("Baseline-v1.csv")
df_baseline_fs["Environment"] = "Baseline-FS"
df_baseline_fscaps = pd.read_csv("Baseline-v1-caps.csv")
df_baseline_fscaps["Environment"] = "Baseline-FSCAPS"

df_enhanced = pd.read_csv("Enhanced-v0.csv")
df_enhanced["Environment"] = "Enhanced"
df_enhanced_caps = pd.read_csv("Enhanced-v0-caps.csv")
df_enhanced_caps["Environment"] = "Enhanced-CAPS"

df_enhanced_fs = pd.read_csv("Enhanced-v1.csv")
df_enhanced_fs["Environment"] = "Enhanced-FS"
df_enhanced_fscaps = pd.read_csv("Enhanced-v1-caps.csv")
df_enhanced_fscaps["Environment"] = "Enhanced-FSCAPS"

df_obstacle = pd.read_csv("Obstacle-v0.csv")
df_obstacle["Environment"] = "Obstacle"
df_obstacle_caps = pd.read_csv("Obstacle-v0-caps.csv")
df_obstacle_caps["Environment"] = "Obstacle-CAPS"

df_obstacle_fs = pd.read_csv("Obstacle-v1.csv")
df_obstacle_fs["Environment"] = "Obstacle-FS"
df_obstacle_fscaps = pd.read_csv("Obstacle-v1-caps.csv")
df_obstacle_fscaps["Environment"] = "Obstacle-FSCAPS"
```

```python
df = pd.concat([df_baseline, df_baseline_caps, df_baseline_fs, df_baseline_fscaps, df_enhanced, df_enhanced_caps, df_enhanced_fs, df_enhanced_fscaps, df_obstacle, df_obstacle_caps, df_obstacle_fs, df_obstacle_fscaps])
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df[df["Episode Length"] != 1]
```

## Part 1

Produce a table in the following format:
|            |             | Episode Length |         |         | CPAD |         |         |
|:----------:|:-----------:|:--------------:|:-------:|:-------:|:----:|:-------:|:-------:|
|            | Environment |       IQR      | Maximum | Minimum |  IQR | Maximum | Minimum |
|    Pure    |   Baseline  |                |         |         |      |         |         |
|            |   Enhanced  |                |         |         |      |         |         |
| Frame Skip |   Baseline  |                |         |         |      |         |         |
|            |   Enhanced  |                |         |         |      |         |         |
|    CAPS    |   Baseline  |                |         |         |      |         |         |
|            |   Enhanced  |                |         |         |      |         |         |
|   FSCAPS   |   Baseline  |                |         |         |      |         |         |
|            |   Enhanced  |                |         |         |      |         |         |

```python
df_no_obstacle = df[df["Environment"].str.contains("Obstacle") == False]
EL_iqr = df_no_obstacle.groupby("Environment")["Episode Length"].quantile([0.25, 0.5, 0.75]).unstack()
EL_max_min = df_no_obstacle.groupby("Environment")["Episode Length"].agg(["max", "min"])
df_el = pd.concat([EL_iqr, EL_max_min], axis=1)
```

```python
grouped_cpad = df_no_obstacle.groupby("Environment")["Cumulative Pairwise Action Distance"]
CPAD_iqr = grouped_cpad.quantile([0.25, 0.5, 0.75]).unstack()
CPAD_max_min = grouped_cpad.agg(["max", "min"])
df_cpad = pd.concat([CPAD_iqr, CPAD_max_min], axis=1)
```

```python
df_obstacle = df[df["Environment"].str.contains("Obstacle")]
sub_df = df_obstacle[df_obstacle["Hit Obstacle"] == 0]
EL_iqr = sub_df.groupby("Environment")["Episode Length"].quantile([0.25, 0.5, 0.75]).unstack()
EL_max_min = sub_df.groupby("Environment")["Episode Length"].agg(["max", "min"])
df_el = pd.concat([EL_iqr, EL_max_min], axis=1)
grouped_cpad = sub_df.groupby("Environment")["Cumulative Pairwise Action Distance"]
CPAD_iqr = grouped_cpad.quantile([0.25, 0.5, 0.75]).unstack()
CPAD_max_min = grouped_cpad.agg(["max", "min"])
df_cpad = pd.concat([CPAD_iqr, CPAD_max_min], axis=1)
```
